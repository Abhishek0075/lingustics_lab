```python
import re


def tokenize(text):
    # Compiling the regex pattern with capturing groups for each part
    pattern = r"""
        (?P<abbreviation>(?:[A-Z]\.)+)                          # Capture group for abbreviations (e.g., U.S.A., U.K.)
        | (?P<contractions>(?:[A-Za-z]+(?:'[A-Za-z]+)+))        # Capture group for contractions (e.g., "isn't")
        | (?P<hyphenated>[A-Za-z]+(?:-[A-Za-z]+)+)              # Hyphenated words (e.g., ice-cream)
        | (?P<word>[A-Za-z]+)                                   # Capture group for regular words
        | (?P<numbers>\d+)                                      # Numbers (whole numbers)
        | (?P<punctuation>[.,!?;:])                             # Punctuation
        | (?P<brackets>[\(\)\[\]{}])                            # Brackets and braces
        | (?P<quotes>["'`]+)                                    # Quotes and other single-character symbols
    """

    # Compile the pattern
    compiled_pattern = re.compile(pattern, re.VERBOSE)

    # Find all matches
    matches = compiled_pattern.finditer(text)

    # List to hold matches with their type and value
    tokens = []

    # Process the matches
    for match in matches:
        
        if match.group('contractions'):
            # Split the contraction to keep it structured
            splitted = match.group('contractions').split("'")
            splitted[1] = splitted[0][-1] + "'" + splitted[1]  # Adjust split for contraction
            splitted[0] = splitted[0][:-1]
            tokens.append(splitted[0])
            tokens.append(splitted[1])
        else:
            # Use a list of token types to iterate through
            token_types = ['abbreviation', 'hyphenated', 'word', 'numbers', 'punctuation', 'brackets', 'quotes']
            for token_type in token_types:
                if match.group(token_type):
                    tokens.append(match.group(token_type))   

    return tokens


sentence = "I love ice-cream ,but he doesn't like it. If he isn't here, I will go to the U.S.A. in 10 hours (from now)."
tokens = tokenize(sentence)
print(tokens)
```

```python
def is_plural_noun_accepted(word):
    vowels = "aeiou"
    
    # Initial state
    if not word or word[-1] != 'y':
        return False  # If it doesn't end with 'y', reject immediately
    
    # Check the second last character
    second_last_char = word[-2] if len(word) > 1 else ''
    
    if second_last_char in vowels:
        # Last character is 'y', and second last is a vowel, accepted as plural
        return True
    elif second_last_char.isalpha() and second_last_char not in vowels:
        # Last character is 'y', and second last is a consonant, accepted as plural
        return True
    else:
        # Any other case is rejected
        return False


# Test cases
test_words = ["boys", "toys", "ponies", "skies", "puppies", "boies", "toies", "ponys"]
results = {word: is_plural_noun_accepted(word) for word in test_words}
print(results)
```
